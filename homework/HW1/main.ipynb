# 3. Import the dataset “Grade.csv” -> Visualization and Analysis

read the dataset Grade.csv

```python
import os
import pandas as pd

GradePath = os.path.join('Analysis', 'data', 'Grade.csv')
GradeData = pd.read_csv(GradePath)
GradeData
```

All missed homework will be scored as zero.

```python
GradeData.columns = GradeData.columns.str.strip()
GradeData.replace(np.nan, 0, inplace=True); 
GradeData.rename(columns={'Mid2':'Midterm 2', 'Homework3':'Homework 3', 'Homework4':'Homework 4'}, inplace=True) # rename

print(GradeData.columns.tolist())
```

```python
homeworkALL = GradeData.drop(columns=['Name', 'Midterm 1','Midterm 2','Final','Participation'])
homeworkAVG = (homeworkALL.sum(axis=1) - homeworkALL.min(axis=1)) / 4
Midterm1 = (GradeData['Midterm 1'])
Midterm2 = (GradeData['Midterm 2'])
Final = (GradeData['Final'])
highest = GradeData[['Midterm 1', 'Midterm 2', 'Final']].max(axis=1)
scores = np.ceil(homeworkAVG * 0.1 + (Midterm1 + Midterm2 + Final)*0.25 + highest * 0.1 + GradeData['Participation'] * 0.05)

scores[(scores>= 57) & (scores < 60)] = 60

perfectAttendanceScores = scores[GradeData['Participation'] == 100]
imperfectAttendanceScores = scores[GradeData['Participation'] != 100]
```

### Visualization: 

```python
from matplotlib.pyplot import subplot
import matplotlib.pyplot as plt
fig, axes = plt.subplots(ncols=2, figsize=(6,8))

edges = np.linspace(0,100,11)
perfectAttendanceScores.plot.hist(bins=edges, ax=axes[0])
axes[0].set_title('Perfect Attendance')
axes[0].set_ylabel('Number of students')
axes[0].set_xlabel('Scores')


imperfectAttendanceScores.plot.hist(bins=edges, ax=axes[1])
axes[1].set_title('Imperfect Attendance')
axes[1].set_ylabel('Number of students')
axes[1].set_xlabel('Scores');
```

### Analysis: 

```python
print(f"The average of those with perfect attendence is {perfectAttendanceScores.mean():.2f}")
print(f"The average of those with imperfect attendence is {imperfectAttendanceScores.mean():.2f}")
corrcoef = np.corrcoef(scores, GradeData['Participation']).round(2)
print(f"Professor Yu collected data for the attendence and the score\nfor 68 students in his class. He found the Pearson correlation\ncoefficient between the two variables to be {corrcoef[0,1]} with\ncorresponding p-value <.001")
```
